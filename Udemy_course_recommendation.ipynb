```python
def search_courses_by_subject(data1, subject):
    # Filter dataset based on subject
    filtered_df = data1[data1['subject'] == subject]

    # Extract relevant columns
    results = filtered_df[['course_title', 'url', 'price','num_subscribers']]

    return results
```

```python
search_courses_by_subject(data1,'Web Development')
```
